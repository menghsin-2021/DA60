### 2. 請將 Dcard API 取得所有的看板資訊轉換成 DataFrame，並且依照熱門程度排序後存成一個 csv 的檔案。


```python

from pprint import pprint

import requests
r = requests.get('https://www.dcard.tw/_api/forums')
response = r.text

import json
data = json.loads(response)  # 清單中裝著字典的資料，字典中有31對key and value
pprint(len(data))  # 485筆字典
pprint(len(data[0]))  # 31對 key and value 於字典中
data_ordered = sorted(data, key=lambda data: data['subscriptionCount'], reverse=True)
# pprint(data_ordered[0].keys())  # 依照 subscriptionCount 排序

df = pd.DataFrame(data_ordered, columns=data_ordered[0].keys())
# print(df)
df.to_csv('output.csv', encoding='utf-8')
pd.read_csv('output.csv')



# pprint(data[0].items())
# pprint(data[1].items())
# data_ordered = sorted(data.items(), key=lambda x: len(x[1]), reverse=True)
#     for d in data_ordered:
#         print(f'{d[0]} 總留言數{len(d[1])}')

```
