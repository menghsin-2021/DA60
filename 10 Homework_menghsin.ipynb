````python
# - select the first 3 rows
print(df.head(3), '\n')

# - select the odd rows. (index = 1, 3, 5)
print(df.iloc[[1, 3, 5], [0, 1, 2, 3]], '\n')

# - select the last 2 columns.
print(df.tail(2), '\n')

# - select the even columns. (index = 0, 2, 4)

print(df.iloc[:, 0], '\n')
print(df.iloc[:, 2], '\n')

print(pd.concat([df.iloc[:, 0],df.iloc[:, 2]],axis=1))
````

### 3. 請根據提供的資料，選擇出下列的要求：

```
# - 1. filtered by first column > 20?
# - 2. filtered by first column + second column > 50
# - 3. filtered by first column < 30 or second column > 30
# - 4. filtered by total sum of row > 100
```

````python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df
````

````python
print(df.loc[:, 0:3][df[0] > 20], '\n')
print(df.loc[:, 0:3][df[0]+df[1] > 50], '\n')
print(df.loc[:, 0:3][(df[0] < 30) & (df[1] > 30)], '\n')
print(df.loc[:, 0:3][df[0]+df[1]+df[2]+df[3] > 100], '\n')
````

````python
# - 1. filtered by first column > 20?
print((df[0] > 20), '\n')
# - 2. filtered by first column + second column > 50
print((df[0]+df[1] > 50), '\n')
# - 3. filtered by first column < 30 or second column > 30
print((df[0] < 30) & (df[1] > 30), '\n')
# - 4. filtered by total sum of row > 100
print((df[0]+df[1]+df[2]+df[3] > 100), '\n')

for i in range(14):
    print(df.iloc[i, :].sum())


````
