```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
df1 = pd.DataFrame(df, columns=['NOX', 'DIS'])
print(df1)
corr = df1.corr(method ='pearson' , min_periods = 1 )
print(corr)

df.plot.scatter(x = 'NOX', y = 'DIS')

#成反比關係 (負相關)
```
